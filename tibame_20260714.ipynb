```python
n = 7
max = n

print("假設 n＝" + str(n))

while True:
    print(n, end="，")
    if n > max:
        max = n
    if n == 1:
        print("迴圈結束，回到 1 了")
        break

    if n % 2 == 0:
        n = n // 2
    else:
        n = 3 * n + 1

print("最高升到：" + str(max))
```
